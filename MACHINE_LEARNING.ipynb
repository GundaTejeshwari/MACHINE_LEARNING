OPERATIONS ON MATRIX


```python
#Matrix addition and subtraction
matrix1 = np.array([[1,2,3],
                    [4,5,6]])
matrix2 = np.array([[1,2,3],
                    [4,5,6]])
print(matrix1+matrix2)
print(matrix1-matrix2)
```

```python
matrix1
```

```python
matrix2
```

```python
#SCALAR MULTIPLICATION
a = np.array([[1,2],
             [3,4]])
sclar = 10
print(scalar*a)
```

```python
#DOT PRODUCT OF VECTORS
# @ and dot() is used to calculate the dot product
vec1 = np.array([3,4])
vec2 = np.array([1,2])
print(vec1 @ vec2)
print(np.dot(vec1,vec2))
```

```python
#DOT PRODUCT ON MULTIPLICATION
mat1 = np.array([[1,2],
             [3,4]])
mat2 = np.array([[1,2],
             [3,4]])
print(np.dot(mat1,mat2))
```

```python
#TRANSPOSE OF A MATRIX
mat1 = np.array([[1,2],
             [3,4]])
print(mat1.T)
```

```python
#DETERMINANT OF MATRIX
mat1 = np.array([[1,2],
             [3,4]])
print(np.linalg.det(mat1))
```

```python
#EIGEN VALUE AND EIGEN VECTOR
eig_val,eig_vec =np.linalg.eig(mat1)
print(eig_val)

print(eig_vec)
```

```python
#INVERSE OF A MATRIX
mat1 = np.array([[1,2],
             [3,4]])
print(np.linalg.inv(mat1))
```
